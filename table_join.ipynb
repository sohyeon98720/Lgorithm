```python
import numpy as np
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori
# from mlxtend.frequent_patterns import association_rules

tbl_demo = pd.read_csv('./LPOINT_BIG_COMP_01_DEMO.csv') # 고객정보
tbl_pdde = pd.read_csv('./LPOINT_BIG_COMP_02_PDDE.csv') # 상품 구매 정보: 유통사 상품 구매 내역
tbl_cop_u = pd.read_csv('./LPOINT_BIG_COMP_03_COP_U.csv') # 제휴사 이용 정보: 제휴사 서비스 이용 내역
tbl_pd_clac = pd.read_csv('./LPOINT_BIG_COMP_04_PD_CLAC.csv') # 상품 분류 정보: 유통사 상품 카테고리 마스터
tbl_br = pd.read_csv('./LPOINT_BIG_COMP_05_BR.csv') # 점포 정보: 유통사/제휴사 점포 마스터
tbl_lpay = pd.read_csv('./LPOINT_BIG_COMP_06_LPAY.csv') # 엘페이 이용: 엘페이 결제 내역(pdde, cop_u와 중복 가능)

cust_id = 'M035502859'
lower_bound = np.percentile(list(tbl_pdde.groupby(['cust']).count()['rct_no']),25) # 하위 25%에 해당하는 구매 내역 수: 30

def most_likely(cust_id):
    # 고객 인적사항 및 구매이력 수 출력
    print(tbl_demo[tbl_demo.cust == cust_id][['ma_fem_dv','ages','zon_hlv']])
    print('A: ',len(tbl_pdde[tbl_pdde.cust==cust_id]))
    print('B: ', len(tbl_cop_u[(tbl_cop_u.cust==cust_id) & (tbl_cop_u['cop_c'].str.contains('B'))]))
    print('C: ', len(tbl_cop_u[(tbl_cop_u.cust==cust_id) & (tbl_cop_u['cop_c'].str.contains('C'))]))
    print('D: ', len(tbl_cop_u[(tbl_cop_u.cust==cust_id) & (tbl_cop_u['cop_c'].str.contains('D'))]))
    print('E: ', len(tbl_cop_u[(tbl_cop_u.cust==cust_id) & (tbl_cop_u['cop_c'].str.contains('E'))]))
    print('Am I satisfy lower_bound?:', 'yes' if len(tbl_pdde[tbl_pdde.cust==cust_id]) >= lower_bound else 'no')
    print()

    # A 기준 구매이력 기반 장바구니 알고리즘(소분류) -> 수정: 구매수량까지 반영
    cust_tmp = pd.merge(left=tbl_pdde[tbl_pdde.cust==cust_id], right=tbl_pd_clac, on='pd_c')
    list_tmp = []
    for i in cust_tmp.index:
        if cust_tmp['buy_ct'][i] >=2:
            for j in range(cust_tmp['buy_ct'][i]):
                list_tmp.append(cust_tmp.iloc[i])
    cust_tmp = cust_tmp.append(list_tmp, ignore_index = True)
    # 구매수량 모두 1로 초기화해야되지만 이제 쓸데없으므로 굳이 안함
    list_pd = cust_tmp.groupby('rct_no')['pd_nm'].apply(list)
    dataset = list(list_pd)
    
    te = TransactionEncoder()
    te_ary = te.fit(dataset).transform(dataset)
    df = pd.DataFrame(te_ary, columns=te.columns_)

    frequent_itemsets = apriori(df,min_support=0.01,use_colnames=True)
    frequent_itemsets.sort_values(by=['support'],axis=0,ascending=False,inplace=True)
    tmp = frequent_itemsets[:5]['itemsets'].values.tolist()
    # print(frequent_itemsets[:5].values.tolist())
    # print(tmp)
    print(sum([list(x) for x in tmp],[]))
    #print(association_rules(frequent_itemsets, metric="confidence", min_threshold=0.01))

most_likely(cust_id)
```

```python
import numpy as np
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori
# from mlxtend.frequent_patterns import association_rules

tbl_demo = pd.read_csv('./LPOINT_BIG_COMP_01_DEMO.csv') # 고객정보
tbl_pdde = pd.read_csv('./LPOINT_BIG_COMP_02_PDDE.csv') # 상품 구매 정보: 유통사 상품 구매 내역
tbl_cop_u = pd.read_csv('./LPOINT_BIG_COMP_03_COP_U.csv') # 제휴사 이용 정보: 제휴사 서비스 이용 내역
tbl_pd_clac = pd.read_csv('./LPOINT_BIG_COMP_04_PD_CLAC.csv') # 상품 분류 정보: 유통사 상품 카테고리 마스터
tbl_br = pd.read_csv('./LPOINT_BIG_COMP_05_BR.csv') # 점포 정보: 유통사/제휴사 점포 마스터
tbl_lpay = pd.read_csv('./LPOINT_BIG_COMP_06_LPAY.csv') # 엘페이 이용: 엘페이 결제 내역(pdde, cop_u와 중복 가능)
lower_bound = np.percentile(list(tbl_pdde.groupby(['cust']).count()['rct_no']),25) # 하위 25%에 해당하는 구매 내역 수: 30

def most_likely(cust_id):
    # 고객 인적사항 및 구매이력 수 출력
    print(tbl_demo[tbl_demo.cust == cust_id][['ma_fem_dv','ages','zon_hlv']])
    print('A: ',len(tbl_pdde[tbl_pdde.cust==cust_id]))
    print('B: ', len(tbl_cop_u[(tbl_cop_u.cust==cust_id) & (tbl_cop_u['cop_c'].str.contains('B'))]))
    print('C: ', len(tbl_cop_u[(tbl_cop_u.cust==cust_id) & (tbl_cop_u['cop_c'].str.contains('C'))]))
    print('D: ', len(tbl_cop_u[(tbl_cop_u.cust==cust_id) & (tbl_cop_u['cop_c'].str.contains('D'))]))
    print('E: ', len(tbl_cop_u[(tbl_cop_u.cust==cust_id) & (tbl_cop_u['cop_c'].str.contains('E'))]))
    print('Am I satisfy lower_bound?:', 'yes' if len(tbl_pdde[tbl_pdde.cust==cust_id]) >= lower_bound else 'no')
    print()

    # A 기준 구매이력 기반 장바구니 알고리즘(소분류) -> 수정: 구매수량까지 반영
    cust_tmp = pd.merge(left=tbl_pdde[tbl_pdde.cust==cust_id], right=tbl_pd_clac, on='pd_c')
    list_tmp = []
    for i in cust_tmp.index:
        if cust_tmp['buy_ct'][i] >=2:
            for j in range(cust_tmp['buy_ct'][i]):
                list_tmp.append(cust_tmp.iloc[i])
    cust_tmp = cust_tmp.append(list_tmp, ignore_index = True)
    # 구매수량 모두 1로 초기화해야되지만 이제 쓸데없으므로 굳이 안함
    list_pd = cust_tmp.groupby('rct_no')['pd_nm'].apply(list)
    dataset = list(list_pd)
    
    te = TransactionEncoder()
    te_ary = te.fit(dataset).transform(dataset)
    df = pd.DataFrame(te_ary, columns=te.columns_)

    frequent_itemsets = apriori(df,min_support=0.01,use_colnames=True)
    frequent_itemsets.sort_values(by=['support'],axis=0,ascending=False,inplace=True)
    print(frequent_itemsets[:5])
    tmp = frequent_itemsets[:5]['itemsets'].values.tolist()
    # print(frequent_itemsets[:5].values.tolist())
    print(sum([list(x) for x in tmp],[]))
    #print(association_rules(frequent_itemsets, metric="confidence", min_threshold=0.01))
```
